# การสร้างกราฟเส้นหลายเส้นด้วย pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter # for custom number format on x and y axis
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
print(f'pandas version: {pd.__version__}')
```

```python
pd.set_option('display.max_rows', 10)
```

```python
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/oil_gold_setindex.csv',
                index_col='Date', parse_dates=True)
df
```

```python
df.plot(kind='line');
```

### matplotlib color name
https://matplotlib.org/2.0.0/examples/color/named_colors.html

```python
df[['Gold', 'SET']].plot(kind='line', color=['gold', 'c']);
```

```python
df['Oil'].plot(kind='line', color='.6', title='OPEC Oil price US$/barrel');
```

### subplots

```python
df[['Oil', 'Gold', 'SET']].plot(kind='line', 
                                subplots=True, 
                                color=['.6', 'gold', 'c']);
```

### figsize

```python
df[['Oil', 'Gold', 'SET']].plot(kind='line', 
                                subplots=True, 
                                figsize=(5, 10),
                                color=['.6', 'gold', 'c']);
```

### layout

```python
df[['Oil', 'Gold', 'SET']].plot(kind='line', 
                                subplots=True, 
                                layout=(3, 1),
                                figsize=(10, 10),
                                color=['.6', 'gold', 'c']);
```

```python
df[['Oil', 'Gold', 'SET']].plot(kind='line', 
                                subplots=True, 
                                layout=(1, 3),
                                figsize=(14, 4),
                                color=['.6', 'gold', 'c']);
```

### adv subplots

```python
fig, ax = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(12, 4))
s = ['Oil', 'Gold', 'SET']
title = ['OPEC Oil price US$/barrel', 'Gold US$/troy ounce', 'SET index']
color = ['.4', 'gold', 'c']
plt.tight_layout()
for i in range(len(s)):
    ax0=df.loc['2003-01-31':, s[i]].plot(kind='line', color=color[i], ax=ax[i], title=title[i])
    ax0.axhline(df[s[i]].mean(), color="red", linestyle="--", linewidth=.8)
    ax0.set_xlabel('');
    ax0.set_ylabel('');    
    ax0.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}".format(v)))
```

### log scale

```python
df.plot(kind='line', color=['.6', 'gold', 'c']);
```

```python
df.plot(kind='line', logy=True, color=['.6', 'gold', 'c']);
```

```python
fig, ax = plt.subplots(1, 2, sharex=False, sharey=False, figsize=(12, 4))
df.plot(kind='line', logy=False, color=['.6', 'gold', 'c'], ax=ax[0])
df.plot(kind='line', logy=True, color=['.6', 'gold', 'c'], ax=ax[1]);
```
